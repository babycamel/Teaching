```python
plt.plot(x,y,label="$y=-0.000023 x^3 + 0.0042 x^2 + 0.75 x$")
plt.xlabel('input')
plt.ylabel('output')
plt.title('Production function')
plt.legend(loc=0)
```

#Technologist's dilemma

Scientists often orient their efforts to increasing productivity, maximising output, e.g. milk yield based on some decision variable.  However the the question is should they aim to maximise total output, incremental (marginal) output or output per unit (average) output.

Marginal output is usually called marginal (physical) product.



# Marginal (Physical) Product

$MPP = \frac{\Delta Y}{\Delta X}$ measures the change in output due to a change in input it can be thought of as the slope of the total output curve between two points. 

I can also be measure in a single point as the slope in that point in this case it is the derivative of the production function $y=f(x)$ or $MPP=\frac{dy}{dx}$.

Using SymPy we can calculate this derivative:

```python
x = Symbol('x') #first define x as a symbol
MPP=diff(-0.000023*x**3 + 0.0042*x**2 + 0.75*x,x) #Calculates derivative symbolically
```

```python
MPP
```

```python
plot(MPP,(x,0,130),title='MPP',xlabel='input',ylabel='litres')
```

or it can be plotted without using the sympy plot function using matplotlib

```python
x = np.linspace(0,130,10)
y = -6.9e-5*x**2 + 0.0084*x + 0.75 #needed to hand copy this from previous cell.
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('MPP')
plt.title('Marginal Physical product')
```

This shows how for each additional unit of inputs, say cows, the marginal output first increases to a peak before declining.

#Inflection point

The inflection point is the point on the total product curve (production function) where the change in the slope changes sign. It corresponds to the maxmimum of the MPP curve. We can find it by maximizing MPP and solving for $X$.

```python
x=Symbol('x')
MPP=-6.9e-5*x**2 + 0.0084*x + 0.75
MPP
secondderiv=diff(MPP,x) #differentiate MPP
inflect=solve(secondderiv,x) #solve for x
inflect #return the infelction point
```

So back to the technologists dilemna what level of inputs is appropriate? Attempting to maximise total product, marginal product and average product simultaneously is clearly not possible. Total product is maximized at an input level above the point where average product is maximised and marginal product is maximised at an input level below that.

```python
x=Symbol('x')
eqn=MPP-APP
print "intersection point:"
intersect=solve(eqn,x)
print intersect
```

#Maximum production

Retuning now to  the total product curve $y=-0.000023 x^3 + 0.0042 x^2 + 0.75 x$. Let's find it's maximum: 

```python
Symbol('x')
y=-0.000023*x**3 + 0.0042*x**2 + 0.75*x
foc=diff(y,x)
print "x that maximizes output"
ans=solve(foc,x)
ans[1]
```

#Brief summary

The three numbers we have solved for divide the space inputs into three sections (F. Knight, Risk, Uncertainty and Profit, 1921, p. 100 https://archive.org/stream/riskuncertaintyp00knigrich#page/98/mode/2up):

- below the point where average product is maximized  is uneconomical (zone I)

- between the point where average product is maximized and the output maximum is economical (zone II)

- above the ouput maximum is uneconomical (zone III)

# The costs of production

We turn now to the total cost of production, marginal cost of production and the average cost of production. 

##Economic costs vs accounting costs

Economic costs and accounting costs are not the same. 

## Total costs

- Total costs are divided into fixed and variable costs of production.

- fixed costs do NOT vary with the level of OUTPUT.

- variable costs vary with the level of OUTPUT.

We refer to total fixed costs and total variable costs.

## Opportunity costs
The cost of the best foregone alternative


## Example
If an animal dies the opportunity cost is the value of all future production from that animal not just the market value of the animal


## Fixed and Variable Costs


Fixed costs are independent of output (y), they do not vary with the output.

Vary with the output (this is why they are called variable costs)

##Total costs
Total costs=Fixed costs (F) + Variable costs (VC)
$$C(y)=F+c(y)$$

Note: variable costs and marginal costs are not the same

#Average Costs

Average Total Costs (ATC)=Average Variable Costs (AVC) +Average Fixed Costs (AFC)
 
ATC=AVC+AFC

$$\frac{C(y)}{y}=\frac{c(y)}{y}+\frac{F}{y}$$
